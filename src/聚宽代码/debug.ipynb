```python
import pandas as pd
import numpy as np
import plotly
```

```python
from dateutil.parser import parse
from types import SimpleNamespace

dict_keywords = {'open_long': '【多头开仓】', 
                'open_short': '【空头开仓】',
                }

# really should use regex, but fuck it 
def check_keyword_type(s):
    for k, v in dict_keywords.items():
        if v in s:
            return k
    return None

def parse_keyword_line(line):
    date, info = line.split('- ERROR -')
    date = parse(date)
    info = info.split('】')[1]
    info = f"SimpleNamespace({info.strip()})"
    info = eval(info)
    return date, info

def parse_line(line):
    keyword_type = check_keyword_type(line)
    if keyword_type is not None:
        date, info = parse_keyword_line(line)
        return (keyword_type, date, info)
    else:
        return None
```

```python
log_file = './log.txt'

logs = []

with open(log_file, 'r') as f:
    for line in f:
        info = parse_line(line)
        if info is not None:
            logs.append(info)

df = pd.DataFrame(logs, columns=['type', 'date', 'info'])
df = df.set_index('date')
df.loc['2021-10-08']
```
